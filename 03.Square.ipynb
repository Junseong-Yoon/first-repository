```python
#### 클래스 만들기 
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()이고,' + self.rect())

        elif self.square == 2:
            print('평행사변형 함수는 par()이고,' + self.par())
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()이고,' + self.trape())
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        width, vertical = map(int, input('가로,세로 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

# 사다리꼴 메서드
    def trape(self):
        width1, width2, vertical = map(int, input('윗변가로,아랫변가로,높이 입력하세요.\n >>>').split(','))
        area = (width1 + width2)*vertical / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result
            
A = Square()
```
