```python
class Square:
    def __init__(self):  # 생성자__init__의 매개변수의 첫번째는 self여야 합니다. 
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))  # 입력값을 int로 묶어야 합니다.

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')  # Square 클래스에 rect() 메서드를 추가합니다. 

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')  # Square 클래스에 par() 메서드를 추가합니다.  
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')  # Square 클래스에 trape() 메서드를 추가합니다. 
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')
```

```python
class Kiosk:
  # 메뉴판 데이터 (클래스 변수)

  opt = ('hot', 'ice')

  coffee = ['americano', 'latte', 'mocha']
  c_price = [2000, 3000, 3000]

  non_coffee = ['yuza_tea', 'green_tea', 'choco_latte']
  n_price = [2500, 2500, 3000]

  korean_menu = {
    'americano': '아메리카노',
    'latte': '라떼',
    'mocha': '카페모카',
    'yuza_tea': '유자티',
    'green_tea': '녹차',
    'choco_latte': '초코라떼'
  }

  def __init__(self):
    self.menus = [(name, price, self.opt)
                  for name, price in zip(self.coffee + self.non_coffee,
                                          self.c_price + self.n_price)]
    self.orders = []

  # 메뉴판 출력
  def show_menu(self):
    base_menu = [(id, (name, price, opt))
                for id, (name, price, opt) in enumerate(self.menus, start=1)]

    for drink_id, (name, price, opt) in base_menu:
        print(f'{drink_id}번. 메뉴 이름: {self.korean_menu.get(name, name)} | 가격: {price} | {opt}\n--------------------------------------------\n')

    return base_menu


  # 메뉴 선택
  def select_menu(self):
    # 메뉴판 데이터 가져오기
    menu = self.show_menu()
    menu_map = {drink_id: (name, price, opt) for drink_id, (name, price, opt) in menu}

    try:
      ordered_id = int(input('음료 번호를 골라주세요: '))

      # 메뉴판에 있는 음료 번호를 입력하지 못했을 시 ValueError를 raise 한다
      if ordered_id not in menu_map:
        raise ValueError


      # hot/ice 입력 아닐 시 ValueError를 raise 한다
      ordered_opt = input('옵션을 골라주세요: 예시: "hot", "ice"')
      if(ordered_opt != 'hot' and ordered_opt != 'ice'):
        raise ValueError

      count = int(input('수량 선택: '))
      if count <= 0:
        raise ValueError('수량은 1 이상이어야 합니다.')

      # 검증 통과
      name, price, _ = menu_map[ordered_id]

      # 주문 누적
      ordered_menu = {
          "id": ordered_id,
          "name": self.korean_menu.get(name),
          "opt": ordered_opt,
          "price": price,
          "count": count,
          "total": price * count,
      }

      self.orders.append(ordered_menu)

      # 주문 정보 리턴
      return ordered_menu

    except ValueError:
      print(f'메뉴에 없는 입력값입니다. 아래 메뉴판을 보고 다시 정확한 값을 입력해주세요\n')
      self.show_menu()

  # 주문 내역이 있는지 체크
  def require_orders(func):
    def wrapper(self, *args, **kwargs):
        if not self.orders:
            print('아직 주문된 메뉴가 없습니다. 먼저 주문을 해주세요.')
            return
        return func(self, *args, **kwargs)
    return wrapper

  @require_orders
  # 주문
  def order_menu(self):
    total_price = 0

    for i, order in enumerate(self.orders, start=1):
        print(f'{i}. {order["name"]} ({order["opt"]}) x{order["count"]} → {order["total"]}원')
        total_price += order["total"]

    print(f'총 주문 금액: {total_price}원\n')

  @require_orders
  # 지불
  def pay(self):
    total_amount = sum(order["total"] for order in Kiosk.orders)
    print(f'총 지불 금액: {total_amount}원')

    payment_method = int(input('1. 현금  2. 카드 \n지불 방법을 번호로 선택해주세요: ').strip())

    if payment_method == 1:
        print('현금 결제를 위해 직원을 호출하겠습니다.')
    elif payment_method == 2:
        print('IC칩 방향에 맞게 카드를 꽂아주세요.')
    else:
        print('올바른 지불 방법을 골라주세요.')


  @require_orders
  # 주문서 출력
  def table(self):
    total_sum = sum(order["total"] for order in self.orders)

    print('\n[주문서 출력]')
    print('⟝' + '-' * 30 + '⟞')

    for _ in range(1):
        print('|' + ' ' * 31 + '|')

    for i, order in enumerate(self.orders, start=1):
        line1 = f'| {i}. {order["name"]} ({order["opt"]}) : {order["price"]}원'
        print(line1.ljust(32) + '|')
        line2 = f'| 수량 : {order["count"]}개 → {order["total"]}원'
        print(line2.ljust(32) + '|')
        print('|' + '-' * 31 + '|')

    total_line = f'| 총 합계 금액 : {total_sum}원'
    print(total_line.ljust(32) + '|')

    for _ in range(1):
        print('|' + ' ' * 31 + '|')
    print('⟝' + '-' * 30 + '⟞')

orderByKiosk = Kiosk()
orderByKiosk.order_menu()
orderByKiosk.select_menu()
```
