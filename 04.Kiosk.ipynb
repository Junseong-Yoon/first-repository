```python
#마지막 추가실습은 gpt활용해서 작성하였음.

from datetime import datetime
from wcwidth import wcswidth

# 문자열 폭 맞춰주는 함수
def pad(text, width):
    space = width - wcswidth(text)
    return text + " " * max(space, 0)

# ✅ 데코레이터 정의
def receipt_deco(func):
    def wrapper(*args, **kwargs):
        line = 40   # 테두리 길이
        print("\n⟝" + "-" * line + "⟞")
        print("|" + " " * (line - 1) + "|")
        func(*args, **kwargs)
        print("|" + " " * (line - 1) + "|")
        print("⟝" + "-" * line + "⟞")
        print("\n결제가 완료되었습니다. 이용해 주셔서 감사합니다.\n")
    return wrapper


class Kiosk:
    def __init__(self, menu):
        self.menu = menu
        self.order_list = []  # [[메뉴명, 가격], ...]
        self.total = 0

    # ✅ 메뉴 출력 (enumerate 활용)
    def menu_print(self):
        print("\n[메뉴판]")
        for i, (name, price) in enumerate(self.menu, start=1):
            print(f"{i}. {pad(name, 14)} : {price:,}원")

    # ✅ 주문 선택
    def menu_select(self):
        while True:
            print()
            n = 0
            while n < 1 or n > len(self.menu):
                try:
                    n = int(input("음료 번호를 입력하세요 : "))
                    if 1 <= n <= len(self.menu):
                        menu_name, menu_price = self.menu[n-1]
                    else:
                        print("없는 메뉴입니다. 다시 입력해주세요.")
                except ValueError:
                    print("숫자를 입력해주세요.")

            # 온도 선택
            t = 0
            while t not in (1, 2):
                try:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                except ValueError:
                    print("숫자를 입력해주세요.\n")

            # 주문 리스트에 추가
            order_name = f"{temp} {menu_name}"
            self.order_list.append([order_name, menu_price])
            self.total += menu_price
            print(f"주문 음료: {order_name} ({menu_price:,}원)")

            more = input("다른 음료를 추가하시겠습니까? (y/n): ").lower()
            if more != 'y':
                break

        # 결제 단계로 이동
        self.pay()

    # ✅ 결제 메서드
    def pay(self):
        print("\n결제 방식을 선택하세요.")
        print("1. 현금 (cash)")
        print("2. 카드 (card)")

        while True:
            pay = input("결제 수단을 입력하세요 (1 또는 2, 또는 'cash'/'card'): ").lower()
            if pay in ('1', 'cash'):
                print("직원을 호출하겠습니다.")
                break
            elif pay in ('2', 'card'):
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("다시 결제를 시도해 주세요.\n")

        # 영수증 출력
        self.receipt()

    # ✅ 영수증 (데코레이터 적용)
    @receipt_deco
    def receipt(self):
        line = 40
        print("| [주문 내역]")
        for i, (menu_name, price) in enumerate(self.order_list, start=1):
            name_field = pad(f"{i}. {menu_name}", 22)
            print(f"| {name_field} : {price:>7,}원")

        print("|" + " " * (line - 1) + "|")
        total_line = pad("합계 금액", 22)
        print(f"| {total_line} : {self.total:>7,}원")

        print("|" + " " * (line - 1) + "|")

        # ✅ 주문 시각
        order_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"| 주문 일시 : {order_time} |")


# ✅ 다차원 리스트로 메뉴 구성
menu = [
    ["아메리카노 (americano)", 2000],
    ["카페라떼 (latte)", 3000],
    ["카페모카 (mocha)", 3000],
    ["유자차 (yuza_tea)", 2500],
    ["녹차 (green_tea)", 2500],
    ["초코라떼 (choco_latte)", 3000]
]

# ✅ 실행
cafe = Kiosk(menu)
cafe.menu_print()
cafe.menu_select()
```
